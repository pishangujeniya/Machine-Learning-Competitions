```python
artist_reputation_missing_value = df['Artist Reputation'].median()
height_missing_value  = df["Height"].median()
width_missing_value = df["Width"].median()
weight_missing_value = df["Weight"].median()
material_missing_value = df["Material"].mode()[0] # most occurred value
transport_missing_value = df["Transport"].mode()[0] # most occurred value
remote_location_missing_value = df["Remote Location"].mode()[0] # most occurred value


missing_value_column_names  = ['Artist Reputation',"Height","Width","Weight","Material","Transport","Remote Location"]

def handleMissingValues(df):  

    df["Artist Reputation"] = df["Artist Reputation"].fillna(artist_reputation_missing_value)
    df["Height"] = df["Height"].fillna(height_missing_value)
    df["Width"] = df["Width"].fillna(width_missing_value)
    df["Weight"] = df["Weight"].fillna(weight_missing_value)
    df["Material"] = df["Material"].fillna(material_missing_value)
    df["Transport"] = df["Transport"].fillna(transport_missing_value)
    df["Remote Location"] = df["Remote Location"].fillna(remote_location_missing_value)
    
    return df

df = handleMissingValues(df)

df.isnull().sum()
```

```python
def handleDates(df):
    df['Scheduled Date']  = pd.to_datetime(df['Scheduled Date'],format="%m/%d/%y")
    df['Delivery Date']  = pd.to_datetime(df['Delivery Date'],format="%m/%d/%y")
    
    df['date_difference'] = (df['Delivery Date'] - df['Scheduled Date']).map(lambda x:str(x).split()[0])
    df['date_difference'] = pd.to_numeric(df['date_difference'])
    
    df['is_early'] = np.where(df['date_difference'] > 0, False, True)
    
    df["scheduled_day"] = df["Scheduled Date"].dt.day
    df["scheduled_month"] = df["Scheduled Date"].dt.month
    df["scheduled_year"] = df["Scheduled Date"].dt.year
    
    df["delivery_day"] = df["Delivery Date"].dt.day
    df["delivery_month"] = df["Delivery Date"].dt.month
    df["delivery_year"] = df["Delivery Date"].dt.year
    
    return df

df = handleDates(df)
df.info()
```

```python
def handleExtraFeatures(df):
    
    df['weight_per_shipping'] = df['Weight'] / df['Base Shipping Price']
    df['volume'] = (df['Height']**2) * (df['Width'])
    
    return df
df = handleExtraFeatures(df)
df.head()
```

```python
# df["Cost"] = df["Cost"].abs()
df["Cost"] = log1p(df["Cost"])
```

```python
def labelEncoding(df, onlyTransform: False, label_object = {}):
    categorical_columns = [
        'Material',
        'International',
        'Express Shipment',
        'Installation Included',
        'Transport',
        'Fragile',
        'Customer Information',
        'Remote Location',
        'is_early'
    ]
    for col in categorical_columns:
        labelencoder = None
        if onlyTransform is False:
            labelencoder = LabelEncoder()
            labelencoder.fit(df[col])
            label_object[col] = labelencoder
        else:
            labelencoder = label_object[col]
            
        df[col] = labelencoder.transform(df[col])
    return df, label_object

df, label_object = labelEncoding(df, onlyTransform= False, label_object= {})
df.head()
```

```python
def oneHotEncoding(df, onlyTransform=False, one_hot_object = {}):
    categorical_columns = [
        'Material',
        'Transport',
        'Customer Information',
    ]
    for col in categorical_columns:
        
        col_data  = df[col].values.reshape(-1,1)
        
        if onlyTransform is False:
            onehot_encoder = OneHotEncoder()
            onehot_encoder = onehot_encoder.fit(col_data)
            one_hot_object[col] = onehot_encoder
        
        onehot_encoded = one_hot_object[col].transform(col_data).toarray()
        columns_list = []
        for i in range(onehot_encoded.shape[1]):
            columns_list.append(col+"_"+str(int(i)))
        dfOneHot = pd.DataFrame(onehot_encoded, columns = columns_list) 
        df = pd.concat([df, dfOneHot], axis=1)
        

    df = df.drop(categorical_columns,axis=1)
    return df, one_hot_object


df, one_hot_object = oneHotEncoding(df, onlyTransform=False, one_hot_object = {})
df.head()
```

```python
redundant_columns = ['Customer Id','Artist Name','Scheduled Date','Delivery Date','Customer Location']
redundant_columns = redundant_columns + ['scheduled_day','scheduled_month','scheduled_year','delivery_day','delivery_month','delivery_year']
df = df.drop(redundant_columns,axis=1)
df.head()
```

```python
y = pd.DataFrame(df["Cost"])
df = df.drop("Cost", axis=1)

X = pd.DataFrame(df)
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)


# X_train = X
# y_train = y

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler



scalerY = StandardScaler()
scalerY = scalerY.fit(y_train)
y_train = scalerY.transform(y_train)
y_train = pd.DataFrame(y_train)

y_test = scalerY.transform(y_test)
y_test = pd.DataFrame(y_test)




scalerX = StandardScaler()
scalerX = scalerX.fit(X_train)
X_train = scalerX.transform(X_train)
X_train = pd.DataFrame(X_train)

X_test = scalerX.transform(X_test)
X_test = pd.DataFrame(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Activation,Dropout 
from tensorflow.keras.regularizers import L2

model = Sequential()
model.add(Dense(512,input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(256, kernel_regularizer=L2(0.01),activation='relu'))
model.add(Dense(256, kernel_regularizer=L2(0.01),activation='relu'))
model.add(Dense(128, kernel_regularizer=L2(0.01), activation='relu'))
model.add(Dense(128, kernel_regularizer=L2(0.01),activation='relu'))
model.add(Dense(64, kernel_regularizer=L2(0.01), activation='relu'))
model.add(Dense(64, kernel_regularizer=L2(0.01),activation='relu'))
model.add(Dense(32, kernel_regularizer=L2(0.01), activation='relu'))
model.add(Dense(32, kernel_regularizer=L2(0.01),activation='relu'))
model.add(Dense(16, kernel_regularizer=L2(0.01), activation='relu'))
model.add(Dense(16, kernel_regularizer=L2(0.01),activation='relu'))
model.add(Dense(8, kernel_regularizer=L2(0.01), activation='relu'))
model.add(Dense(8, kernel_regularizer=L2(0.01),activation='relu'))
model.add(Dense(1, kernel_initializer='normal',activation='linear'))

model.compile(optimizer='adam',
                loss='mean_squared_error',
                metrics=['mean_squared_logarithmic_error'])

print(model.summary())
```
